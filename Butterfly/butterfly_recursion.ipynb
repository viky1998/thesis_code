```python
from Butterfly import Butterfly
butterfly = Butterfly()
```

## Recursion on normal Butterfly network

```python
# print the recursive list for a given n
n= 15
recursive_list = butterfly.recursive_list(n)
print(recursive_list)
```

```python
# Find first occurence of c in the recursive_list
c = 6
index = butterfly.find_first_occurrence(recursive_list, c)
print(f"The first occurrence of {c} is at index {index}")
```

```python
# Find the sum before the first occurence of c in the recursive_list
sum_result = butterfly.sum_before_first_occurrence(recursive_list, c)
print(f"The sum of values before the first occurrence of {c} is {sum_result}")
```

```python
# Check when the sum becomes greater than 256 and return the next element
next_element = butterfly.check_sum_and_next_element(recursive_list, c)
print(f"The next element after the sum becomes greater than 256 is {next_element}")
```

```python
# Given a value of c, find the sum of all values before the first occurrence of
# c in the recursive list for n in n_range
c=10
n_range = range(10,26)
df_results = butterfly.get_table_sum_before_c(c, n_range)
df_results = df_results.query('sum_before_first_c > 256')
print(df_results)
```

```python
# Given a range for c and n, find the sum of all values before the first
# occurrence of c in the recursive list
c_range = range(5,20)
n_range = range(8,26)
df_sums = butterfly.get_table_sum_before(c_range, n_range)
df_pivot_sums = df_sums.pivot(index='c', columns='logn', values='sum_before_first_c')
print(df_pivot_sums)
```

## Formula for sum before first occurrence

```python
# For given n and c calculate the sum using the formula and using the recursive list
n=14
c=6
recursion_list = butterfly.recursive_list(n)
index = butterfly.find_first_occurrence(recursion_list,c)
sum = butterfly.sum_before_first_occurrence(recursion_list,c)
sum_formula = butterfly.calc_sum_formula(n, c)

print(f"Recursion_list: {recursion_list}")
print(f"Index of first occurence of c: {index}")
print(f"Sum using reursive list: {sum}")
print(f"Sum using formula: {int(sum_formula)}")
```

```python
# check if the sum is correct for all values of n and c in given ranges
n_range = range(8,20)
c_range = range(1,15)

for n in n_range:
    for c in c_range:
        sum = butterfly.sum_before_first_occurrence(butterfly.recursive_list(n), c)
        sum_formula = butterfly.calc_sum_formula(n,c)
        if c<n: 
            assert sum == int(sum_formula) 
```
