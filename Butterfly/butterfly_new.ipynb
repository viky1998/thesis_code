```python
from butterfly import Butterfly
butterfly = Butterfly()
```

```python
c_range = range(5, 20)
N_range = range(9, 24)
df_sum_before = butterfly.get_table_sum_before(c_range, N_range)
pivot_sum_before = df_sum_before.pivot(index='c', columns='logn', values='sum_before_first_c')
print(pivot_sum_before)
```

```python
pivot_sum_before.to_csv('pivot_sum_before.csv')
```

```python
c_range = range(5, 20)
N_range = range(9, 28)
df_necessary_l = butterfly.get_table_necessary_l(c_range, N_range)
pivot_necessary_l = df_necessary_l.pivot(index='c', columns='logn', values='l')
print(pivot_necessary_l)
```

```python
pivot_necessary_l.to_csv('pivot_necessary_l.csv')
```

```python
c_range = range(5, 20)
N_range = range(18, 28)
df_sum_before_with_l = butterfly.get_table_sum_before_with_l(c_range, N_range)
pivot_sum_before_with_l = df_sum_before_with_l.pivot(index=['c','l'], columns='logn', values='sum_before_first_c')
print(pivot_sum_before_with_l)
```
